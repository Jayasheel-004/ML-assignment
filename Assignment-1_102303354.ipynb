Find the most frequent value and their indice(s) in the following arrays:
i. x = np.array([1,2,3,4,5,1,2,1,1,1])
ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python

x = np.array([1,2,3,4,5,1,2,1,1,1])
most_freq_x = np.bincount(x).argmax()
indices_x = np.where(x == most_freq_x)[0]
print("Most frequent in x:", most_freq_x)
print("Indices:", indices_x)


y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_freq_y = np.bincount(y).argmax()
indices_y = np.where(y == most_freq_y)[0]
print("Most frequent in y:", most_freq_y)
print("Indices:", indices_y)
```

For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), find
i. Sum of all elements
ii. Sum of all elements row-wise
iii. Sum of all elements column-wise

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = gfg.sum()
row_sum = gfg.sum(axis=1)
col_sum = gfg.sum(axis=0)

print("Total sum:", total_sum)
print("Row-wise sum:\n", row_sum)
print("Column-wise sum:\n", col_sum)
```

For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find
i. Sum of diagonal elements
ii. Eigen values of matrix
iii. Eigen vectors of matrix
iv. Inverse of matrix
v. Determinant of matrix

```python
n_array = np.array([[55, 25, 15],
                    [30, 44, 2],
                    [11, 45, 77]])


diag_sum = np.trace(n_array)
eigen_values, eigen_vectors = np.linalg.eig(n_array)
inverse_matrix = np.linalg.inv(n_array)
det = np.linalg.det(n_array)

print("Sum of diagonal elements:", diag_sum)
print("Eigen values:", eigen_values)
print("Eigen vectors:\n", eigen_vectors)
print("Inverse matrix:\n", inverse_matrix)
print("Determinant:", det)
```

Multiply the following matrices and also find covariance between matrices using NumPy:
i. p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]]
ii. p = [[1, 2], [2, 3], [4, 5]]
q = [[4, 5, 1], [6, 7, 2]]

```python

p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])

product1 = np.dot(p1, q1)
cov1 = np.cov(p1.T, q1.T)

print("Product (1):\n", product1)
print("Covariance (1):\n", cov1)

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

product2 = np.dot(p2, q2)
cov2 = np.cov(p2.T, q2[:2].T)  

print("Product (2):\n", product2)
print("Covariance (2):\n", cov2)
```

For the matrices: x = np.array([[2, 3, 4], [3, 2, 9]]); y = np.array([[1, 5, 0], [5, 10, 3]]),
find inner, outer and cartesian product?

```python
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

inner = np.inner(x, y)
outer = np.outer(x.flatten(), y.flatten())
from itertools import product
cartesian = np.array(list(product(x.flatten(), y.flatten())))

print("Inner product:\n", inner)
print("Outer product:\n", outer)
print("Cartesian product:\n", cartesian)
```

For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
array = np.array([[1, -2, 3],
                  [-4, 5, -6]])
```

```python
abs_array = np.abs(array)
print("Absolute values:\n", abs_array)
```

```python
flat = array.flatten()
print("Flattened array percentiles:")
print("25th:", np.percentile(flat, 25))
print("50th:", np.percentile(flat, 50))
print("75th:", np.percentile(flat, 75))
```

```python
print("Column-wise percentiles:")
print("25th:", np.percentile(array, 25, axis=0))
print("50th:", np.percentile(array, 50, axis=0))
print("75th:", np.percentile(array, 75, axis=0))
```

```python
print("Row-wise percentiles:")
print("25th:", np.percentile(array, 25, axis=1))
print("50th:", np.percentile(array, 50, axis=1))
print("75th:", np.percentile(array, 75, axis=1))
```

```python
print("Flattened stats:")
print("Mean:", np.mean(flat))
print("Median:", np.median(flat))
print("Std Dev:", np.std(flat))
```

```python
print("Column-wise stats:")
print("Mean:", np.mean(array, axis=0))
print("Median:", np.median(array, axis=0))
print("Std Dev:", np.std(array, axis=0))
```

```python
print("Row-wise stats:")
print("Mean:", np.mean(array, axis=1))
print("Median:", np.median(array, axis=1))
print("Std Dev:", np.std(array, axis=1))
```

For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]). Find floor, ceiling
and truncated value, rounded values

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor:", np.floor(a))       # Nearest lower integer
print("Ceiling:", np.ceil(a))      # Nearest higher integer
print("Truncated:", np.trunc(a))   # Remove decimal part (towards zero)

print("Rounded (default):", np.round(a))        # Round to nearest integer
print("Rounded (1 decimal):", np.round(a, 1))   # Round to 1 decimal
```

For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print("Sorted array:", sorted_array)


sorted_indices = np.argsort(array)
print("Indices of sorted array:", sorted_indices)


smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)


largest_5 = np.sort(array)[-5:][::-1]
print("5 largest elements:", largest_5)
```

For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
i. Integer elements only
ii. Float elements only

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integers = array[np.equal(np.mod(array, 1), 0)]
print("Integer elements:", integers)

floats = array[np.not_equal(np.mod(array, 1), 0)]
print("Float elements:", floats)
```
